```python
import numpy as np 
import pandas as pd 
import os
import glob
import math
import matplotlib.pyplot as plt 
import seaborn

#to read all the csv placed in a folder

path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))

df=pd.DataFrame()
#cd /Users/utkarshruhela/Downloads/archive
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    x=f[39:]
    y=x[:-4]
    print(y)
    df1 = pd.read_csv(f)
    df1['Symbol']=y
    df=pd.concat([df, df1])
    
#calculating daily return
df['Return']=(df['Close']-df['Prev Close'])/df['Prev Close']
df.replace(np.inf, np.nan)

#calculating price close correlations
df_pivot1=df.pivot('Date','Symbol','Close').reset_index()
df_pivot1.head()

corr_df1 = df_pivot1.corr(method='pearson')
#reset symbol as index (rather than 0-X)

#del corr_df.index.name
corr_df1

df_pivot2=df.pivot('Date','Symbol','Return').reset_index()
df_pivot2.head()

corr_df2 = df_pivot2.corr(method='pearson')
#reset symbol as index (rather than 0-X)

#del corr_df.index.name
corr_df2

corr_df1.to_csv(path+"/ClosingCorrelation.csv")
corr_df2.to_csv(path+"/ReturnCorrelation.csv")

plt.figure(figsize=(30, 15))
seaborn.heatmap(corr_df2, cmap='RdYlGn', annot=True)
plt.savefig('saving-a-seaborn-plot-as-png-file-transparent.png')
```
